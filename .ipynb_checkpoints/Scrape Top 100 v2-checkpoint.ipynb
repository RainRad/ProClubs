```python
import requests
import pandas as pd
import MyFunctionsList as mfs
from DoNotPublish import password
from sqlalchemy import create_engine
from bs4 import BeautifulSoup as soup
import datetime

# Debugging
import time
from IPython.display import display, clear_output
```

```python
# Debug
start_time = time.time()

# Creates a dictionary to fill later. Key = Console, Value = [List of Clubs]
final_multi_console_ids = {}

consoles = ["xboxone", "ps4","xbsxs","ps5","pc"]

# Iterate through each console's top 100
for console in consoles:
    
    console_id_list = []
    # Construct URL, Scrape Website
    url = f"https://proclubshead.com/23/leaderboard-club/{console}"
    site = mfs.scrape(url)
    
    for club in site.find_all("a", class_="btn btn-secondary d-block mb-3 p-0", href=True):
        club_link = club['href']
        club_id = mfs.simple_link_stripper(club_link)
        console_id_list.append(club_id)        
    final_multi_console_ids[console] = console_id_list
    
    
# Debug
end_time = time.time()
print("Time taken:", end_time - start_time, "seconds")
```

```python
# Debug
start_time = time.time()
length  = 0
for console in final_multi_console_ids:
    length += len(final_multi_console_ids[console])
counter = 0

# Iterate through each console. 
final_club_data_list = []


for console in final_multi_console_ids:
    
    #Iterate through each club. 
    for club in final_multi_console_ids[console]:
        print(club)
        club_data = mfs.club_scrape(console, club)
        final_club_data_list.append(club_data)
        
        #debug
        clear_output(wait=True)
        counter += 1
        print(f"{counter}/{length}")
        
# Debug
end_time = time.time()
print("Time taken:", end_time - start_time, "seconds")
```

```python
#Generates and stores current time for later use. 
current_time = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
Top_Clubs = pd.DataFrame(final_club_data_list)

Top_Clubs.insert(0, 'pull_time', current_time)

Top_Clubs.head()
```

```python
protocol = 'postgresql'
username = 'postgres'
password = password
host = 'localhost'
port = 5432
database_name = 'pro_clubs_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
Top_Clubs.to_sql('clubs', engine, if_exists='append', index=False)
```
