```python
sql_query = 'SELECT g.*, g.h_club_id, gc.wins AS h_wins, gc.losses AS h_losses, gc.draws AS h_draws, gc.division AS h_divison,\
gc.ten_match_record as h_tm_record, gc.goals_per_match as h_goals_per_match,\
gc.goals_conceded_per_match as h_goals_conceded_per_match, g.opp_club_id, go.wins AS opp_wins, go.losses AS opp_losses,\
go.draws AS opp_draws, go.division AS opp_division, go.ten_match_record as opp_tm_record,\
go.goals_per_match as opp_goals_per_match, go.goals_conceded_per_match as opp_goals_conceded_per_match \
FROM games g JOIN clubs gc ON g.h_club_id = gc.club_id JOIN clubs go ON g.opp_club_id = go.club_id'
```

```python
df = pd.read_sql(sql_query, engine)
```
