```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
import seaborn as sn
import numpy as np
import sys
sys.path.insert(0, '../Resources')
import MyFunctionsV4 as mfs
```

```python
match_data_link = "../CollectedData/Merged_Club_Match_Data_92820.csv"
data_df = pd.read_csv(match_data_link)
```

```python
pd.options.display.max_columns = None
df = data_df.drop('Unnamed: 0', axis=1)
df.head(15)
df.columns
```

```python
def results(row):
    if row.winning_club == row.h_club:
        row['result'] = 1
    elif row.winning_club == row.opp_club:
        row['result'] = 0
    else:
        row['result'] = 2
    
    count=0
    total=0
    for c in row.h_ten_match_record:
        if c == 'W':
            total += 3
        elif c == 'D':
            total += 1
        elif c == 'L':
            total += 0
        count += 1
        score = total / count
    row.h_ten_match_record = score
    
    count=0
    total=0
    for c in row.opp_ten_match_record:
        if c == 'W':
            total += 3
        elif c == 'D':
            total += 1
        elif c == 'L':
            total += 0
        count += 1
        score = total / count
    row.opp_ten_match_record = score
    return row
```

```python
df = df.apply(results, axis=1)
df.columns
```

```python
df_trimmed = df[[
       'h_division', 'h_ten_match_record', 'h_total_games', 'h_wins',
       'h_losses', 'h_draws', 'h_goals_scored', 'h_goals_per_match',
       'h_goals_conceded', 'h_goals_conceded_per_match',
       'h_total_rank_points', 'h_div_1_titles', 'h_total_trophies',
       'h_goals', 'h_shots', 'h_shot_percent', 'h_passes_made',
       'h_pass_attempts', 'h_pass_percent', 'h_tackles_made',
       'h_tackle_attempts', 'h_tackle_percent', 'h_red_cards',
       'h_players_in_match', 'h_forward', 'h_midfielder', 'h_defender',
       'h_goalkeeper', 'h_any', 'h_season', 'h_round', 'opp_division',
       'opp_ten_match_record', 'opp_total_games', 'opp_wins', 'opp_losses',
       'opp_draws', 'opp_goals_scored', 'opp_goals_per_match',
       'opp_goals_conceded', 'opp_goals_conceded_per_match',
       'opp_total_rank_points', 'opp_div_1_titles', 'opp_total_trophies',
       'opp_goals', 'opp_shots', 'opp_shot_percent', 'opp_passes_made',
       'opp_pass_attempts', 'opp_pass_percent', 'opp_tackles_made',
       'opp_tackle_attempts', 'opp_tackle_percent', 'opp_red_cards',
       'opp_players_in_match', 'opp_forward', 'opp_midfielder', 'opp_defender',
       'opp_goalkeeper', 'opp_any', 'opp_season', 'opp_round', 'result']]
```

```python
print(sum(np.isnan(df_trimmed).sum()))
print(sum(np.isinf(df_trimmed).sum()))

print(len(df_trimmed))
df_trimmed = df_trimmed.replace([np.inf, -np.inf], np.nan)
df_trimmed = df_trimmed.dropna(how='any', subset=None, inplace=False)
print(len(df_trimmed))

print(sum(np.isnan(df_trimmed).sum()))
print(sum(np.isinf(df_trimmed).sum()))
```

```python
X = df_trimmed.drop(['result', 'h_shot_percent', 'h_goals','opp_goals', 'opp_shot_percent'], axis=1)
y = df_trimmed.result
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
classifier = RandomForestClassifier(n_estimators=750, criterion ='entropy', random_state = 1)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
classifier.feature_importances_
```

```python
# Generate bar graph of feature importance. 
importance = classifier.feature_importances_
columns = X_train.columns
combined_features = pd.Series(importance, columns)

# sort the combined_features series in descending order
combined_features_sorted = combined_features.sort_values(ascending=False)

# plot the sorted series
pandas_plot = combined_features_sorted.plot.bar(figsize=(15,6), color='b', fontsize = 12)
plt.xlabel("columns", fontsize = 12)
# plt.xticks(rotation = 45)
plt.ylabel("importance", fontsize = 12)
plt.title("Important Contributors for MOTM", fontsize = 18)
plt.tight_layout()
plt.show()
```
