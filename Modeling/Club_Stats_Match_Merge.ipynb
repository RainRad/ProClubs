```python
matches_file = "../CollectedData/match_list_92820.csv"
clubs_file = "../CollectedData/club_list_47422.csv"
matches_df = pd.read_csv(matches_file)
clubs_df = pd.read_csv(clubs_file)
```

```python
clubs_df['club_id'] = clubs_df['club_id'].astype(int)
```

```python
home_club_match_stats = matches_df[['h_goals', 'h_shots', 'h_shot_percent', 'h_passes_made',
       'h_pass_attempts', 'h_pass_percent', 'h_tackles_made',
       'h_tackle_attempts', 'h_tackle_percent', 'h_red_cards',
       'h_players_in_match', 'h_forward', 'h_midfielder', 'h_defender',
       'h_goalkeeper', 'h_any', 'h_season', 'h_round']]

opp_club_match_stats = matches_df[['opp_goals', 'opp_shots', 'opp_shot_percent',
       'opp_passes_made', 'opp_pass_attempts', 'opp_pass_percent',
       'opp_tackles_made', 'opp_tackle_attempts', 'opp_tackle_percent',
       'opp_red_cards', 'opp_players_in_match', 'opp_forward',
       'opp_midfielder', 'opp_defender', 'opp_goalkeeper', 'opp_any',
       'opp_season', 'opp_round']]

clubs_df_subset = clubs_df[['club_name','club_id', 'division',
       'ten_match_record', 'total_games', 'wins', 'losses', 'draws',
       'goals_scored', 'goals_per_match', 'goals_conceded',
       'goals_conceded_per_match','region','total_rank_points',
       'div_1_titles','total_trophies']].fillna('region')

empty_value_df = pd.DataFrame(columns=['h_division',
       'h_ten_match_record', 'h_total_games', 'h_wins', 'h_losses', 'h_draws',
       'h_goals_scored', 'h_goals_per_match', 'h_goals_conceded',
       'h_goals_conceded_per_match','h_region','h_total_rank_points',
       'h_div_1_titles','h_total_trophies','opp_division',
       'opp_ten_match_record', 'opp_total_games', 'opp_wins', 'opp_losses', 'opp_draws',
       'opp_goals_scored', 'opp_goals_per_match', 'opp_goals_conceded',
       'opp_goals_conceded_per_match','opp_region','opp_total_rank_points',
       'opp_div_1_titles','opp_total_trophies'])
```

```python
def club_match_data_assign(row, clubs_df_subset, home_club_match_stats, opp_club_match_stats):
    h_club_info = clubs_df_subset[clubs_df_subset['club_id'] == row.h_club_id].reset_index(drop=True)
    opp_club_info = clubs_df_subset[clubs_df_subset['club_id'] == row.opp_club_id].reset_index(drop=True)
    row.loc['h_division':'h_total_trophies'] = h_club_info.iloc[0, 2:].values
    row.loc['opp_division':'opp_total_trophies'] = opp_club_info.iloc[0, 2:].values
    return row
```

```python
base_df = pd.concat([matches_df, empty_value_df], axis=1)
```

```python
output_df = base_df.apply(club_match_data_assign, args=(clubs_df_subset, home_club_match_stats, opp_club_match_stats), axis=1)
```
